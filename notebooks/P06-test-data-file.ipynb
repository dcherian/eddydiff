````python
original = xr.open_dataset(path)
original
````

````python
def reformat_ctd_chipod_nc(ds):
    ds = ds.copy(deep=True)
    ds = (
        ds.set_coords(["lat", "time", "CTD-chipod"])
        .rename({"timeseries": "trajectory", "profile": "pressure"})
        .rename({"trajectory": "profile"})
    )

    ds["pressure"] = ds["P"]
    del ds["P"]
    ds["profile"] = (
        "profile",
        ds.station_dn.where(ds.station_dn > 0).interpolate_na("profile").data,
        ds.station_dn.attrs,
    )
    ds.profile.attrs["cf_role"] = "profile_id"

    ds["direction"] = ("direction", ["up", "dn"])

    # fmt: off
    for var in ["T", "S", "SN", "station", "dThdz", "pts2bin", "N2", "chi", 
                "eps", "KT", "chiGE", "epsGE", "KTGE", "GEflag", "sn_avail"]:
    # fmt: on
        with xr.set_options(keep_attrs=True):
            ds[var] = xr.concat([ds[f"{var}_up"], ds[f"{var}_dn"]], dim="direction") 
        ds[var].attrs["_FillValue"] = int(ds[f"{var}_up"].attrs["FillValue"])
        ds = ds.drop_vars([f"{var}_up", f"{var}_dn"])
        del ds[var].attrs["FillValue"]
        
    ds = xr.decode_cf(ds)

    ds["SN"] = ds.SN.astype(int)
    ds["sn_avail"] = ds.sn_avail.astype(int)
    return ds

    ds = reformat_ctd_chipod_nc(original).load()
````
